```python
#Analysis for total data
data = pd.read_csv('data.csv')
cols = ['Q5','Q6','Q7','Q8','Q9','Q10','Q11','Q12','Q13','Q16','Q17','Q18','Q19','Q20','Q21','Q22','Q25', 'Q26', 'Q32', 'Q33', 'Q34', 'Q191']
filtered_data = data[cols]
filtered_data.columns = filtered_data.iloc[0]
filtered_data = filtered_data[2:]

#Loop over data to extract values
response_dict = {}

for i in filtered_data.columns:
    count = filtered_data[i].count()
    agree = np.round(((filtered_data[i].value_counts()['Agree'] + filtered_data[i].value_counts()['Strongly Agree'])/count),2)
    disagree = np.round(((filtered_data[i].value_counts()['Disagree'] + filtered_data[i].value_counts()['Strongly Disagree'])/count),2)
    neither = np.round((filtered_data[i].value_counts()['Neither Agree nor Disagree']/count),2)
    response_dict[i] = (agree, neither, disagree)
    
#Create Dataframe 
response = pd.DataFrame(response_dict).T
response.columns = columns = ['Agree', 'Neither', 'Disagree']
response['strength'] = np.where(response['Agree'] >= .75, 'yes', 'no')
response['improve'] = np.where(response['Neither'] + response['Disagree'] >= .40, 'yes', 'no')
response['question'] = response.index
response['order'] = range(1,len(response['question'])+1,1)
cols = [ 'order','question','Agree', 'Neither', 'Disagree', 'strength', 'improve']
response = response[cols]
response.to_csv('total.csv')
```
